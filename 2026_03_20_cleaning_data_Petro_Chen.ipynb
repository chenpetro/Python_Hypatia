```python
def parse_batch1_dat(path):
    rows = []
    gas_id = None

    with open(path, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            # Gas‑label line: single digit 1–6
            if len(line) == 1 and line in "123456":
                gas_id = int(line)
                continue

            # Otherwise this line is a feature row: 1:... 2:...
            features = {"gas": gas_id}  # attach gas label from previous line
            for part in line.split():
                if ":" not in part:
                    continue  # ignore malformed parts
                idx, val = part.split(":", 1)
                features[int(idx)] = float(val)

            rows.append(features)

    return pd.DataFrame(rows)

df = parse_batch1_dat("batch1.dat")
print(df.shape)
print(df.head())
```
