```python

import numpy as np

# now we show how to do slicing.  the general format for a 2x2 array is

# a[from : to -1 , from : to -1]

# the colon by itself means all elements, like a[:,:]

# so the comma seperates the axis and the colon says what to do within the axis


# first columns from the 4th and 5th rows
# remember we start counting array elements at 0 so we write 3 when we mean element 4

arr=np.random.randint(1,100,(10,10))

arr
```

```python
arr[3:5,:]
```

```python
# arr[rows, columns]

# arrr[start:stop -1, start-stop - 1]

arr[3:5,0:2]
```

```python
import numpy as np
```

```python
#1 (difficult) create a 4 x 4 array, how can you print elements along the diagonal. write a for-loop
arr0 = np.arange(1, 17).reshape(4,4)

print(arr0)

for i in range(arr0.shape[0]):
    print(arr0[i, i])
```
